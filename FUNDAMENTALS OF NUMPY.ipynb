# Array operations


```python
a = np.array([1,2,3,4,5])
b = np.array([5,6,7,8,9])

c = a-b
print('sub',c)

d = a+b
print('add',d)

e = a*b
print('mul',e)
```

```python
fahrenhit = np.array([0,-20,-21.5,-12.3,15,0])
#to convert celsius fahrenhit-32*5/9
celsius = (fahrenhit-31)*5/9
celsius
```

```python
celsius > -20
```

```python
celsius%2 == 0
```

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
#for matrix multiplication we use @ operator
print(a*b)
print(a@b)
```

```python
a.dtype
a.ndim
```

```python
a = np.array([1,2,3])
b = np.array([4.2,5.5,6.23])
c = a+b
print(c)
print(c.dtype)
```

```python
print(c.sum())
print(c.mean())
print(c.min())
print(c.max())
```
